```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("C:\\Users\\Dell\\Videos\\housing.csv")
```

```python
#printing the first 5 rows
data.head()
```

```python
#the info() method is typically used to get a concise summary of a DataFrame, including the data types and non-null counts for each column.
data.info()
```

```python
#The dropna() method is used to remove missing values (NaN or null values) from a DataFrame as we have small number of non values so we will drop them.
data.dropna(inplace=True)
```

```python
data.info()
```

So there are no null values

```python

X=data.drop(['median_house_value'],axis=1)
Y=data['median_house_value']
```

```python
#X is  feature matrix.
#Y is target variable.
#The train_test_split function randomly splits thw data into training and testing sets. The test_size=0.2 parameter indicates that 20% of the data will be used for testing, and the remaining 80% will be used for training.
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
#3joining the training data
train_data=X_train.join(Y_train)
train_data
```

```python
train_data.hist(figsize=(15,8))
```

It looks like data columns are mostly left skewed

```python
#applying the natural logarithm (np.log()) transformation to4 columns in train_data.
#Logarithmic transformations are commonly used in linear regression to handle skewed data or to stabilize variance.
train_data['total_rooms']=np.log(train_data['total_rooms'])
train_data['total_bedrooms']=np.log(train_data['total_bedrooms'])
train_data['population']=np.log(train_data['population'])
train_data['households']=np.log(train_data['households'])
```

```python
train_data.hist(figsize=(15,8))
```

The data is more centered

```python
#one-hot encoding the 'ocean_proximity' column in  train_data and then join it back to  DataFrame while dropping the original column
train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
```

```python
train_data['bedroom_ratio']=train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms']=train_data['total_rooms']/train_data['households']
```

```python
#one-hot encoding the 'ocean_proximity' column in  train_data and then join it back to  DataFrame while dropping the original column
train_data= train_data.join(pd.get_dummies(train_data['ocean_proximity'])).drop(['ocean_proximity'], axis=1)

plt.figure(figsize=(15, 8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
plt.show()
```

we see that when houses are in Land the housevalue and house income are less than when it is island

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='latitude',y='longitude',data=train_data,hue='median_house_value',palette='coolwarm')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_train,Y_train=train_data.drop(['median_house_value'],axis=1),train_data['median_house_value']
#fitting a linear regression model using LinearRegression()
reg=LinearRegression()
#fit the model to the training data using reg.fit(X_train, Y_train) to make predictions
reg.fit(X_train,Y_train)
```

```python
#Now, the test_data  contains the log-transformed features and one-hot encoded 'ocean_proximity' column.
test_data=X_test.join(Y_test)
test_data['total_rooms']=np.log(test_data['total_rooms'])
test_data['total_bedrooms']=np.log(test_data['total_bedrooms'])
test_data['population']=np.log(test_data['population'])
test_data['households']=np.log(test_data['households'])
test_data=test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
```

```python
test_data
```

```python
X_test,Y_test=test_data.drop(['median_house_value'],axis=1),test_data['median_house_value']
```

```python
#The score method in scikit-learn's linear regression model calculates the coefficient of determination, 
#commonly known as R-squared. It measures the proportion of the variance in the dependent variable that is predictable from the independent variables. The score method expects the features (X_test) and the true target values (Y_test).
reg.score(X_test,Y_test)
```

The R-squared score of approximately 0.666 on  test data indicates that linear regression model explains about 66.6% of the variance in the target variable (median_house_value).
